```python
import requests
import pandas as pd

from datetime import date, datetime, timedelta
from src.utils import get_config, Logger

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

config = get_config()
log = Logger()
```

```python
today = date.today()
symbol = 'VZ'
print(f"https://financialmodelingprep.com/api/v3/historical-price-full/{symbol}?from=2021-07-01&to={today}apikey={config['credentials']['financialmodelingprep']}")
```

```python
r = requests.get(f"https://financialmodelingprep.com/api/v3/historical-price-full/{symbol}?from=2021-07-01&to={today}&apikey={config['credentials']['financialmodelingprep']}")
r.json()
df = pd.json_normalize(r.json())
```

```python
history = df['historical'].values[0]
hist_df = pd.json_normalize(history)
display(hist_df)
```

```python
dates = hist_df['date'].values
dates.sort()
a = dates[-1]
```

```python
b = pd.to_datetime(a) + timedelta(days=1)
print(b.date())
```
